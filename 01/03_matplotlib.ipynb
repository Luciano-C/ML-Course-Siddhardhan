```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Generate data to plot
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

Plotting the data

```python
# Sine wave
plt.plot(x,y)
plt.show()
```

```python
# Cosine wave
plt.plot(x,z)
plt.show()
```

```python
# Adding title to x axis and y axis labels
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('Sine Wave')
plt.show()
```

```python
# Parabola
x = np.linspace(-10, 10, 20)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y, 'r+')
plt.show()
```

```python
plt.plot(x,y, 'g.')
plt.show()
```

```python
plt.plot(x,y, 'rx')
plt.show()
```

```python
x = np.linspace(-5, 5, 50)
plt.plot(x, np.sin(x), 'g-')
plt.plot(x, np.cos(x), 'r--')
plt.show()
```

Bar plots

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['English', 'French', 'Spanish', 'Latin', 'German']
people = [100, 50, 150, 40, 70]
ax.bar(languages, people)
plt.xlabel('LANGUAGES')
plt.ylabel('NUMBER OF PEOPLE')
plt.show()
```

Pie chart

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ['English', 'French', 'Spanish', 'Latin', 'German']
people = [100, 50, 150, 40, 70]
# autopct tiene que ver con el formato de los decimales
ax.pie(people, labels=languages, autopct='%1.1f%%')
plt.show()
```

Scatter plot

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)

fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show()
```

3D scatter plot

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
#c: array-like or list of colors or color, optional
ax.scatter(x, y, z, c=z, cmap='Blues')
plt.show()
```
